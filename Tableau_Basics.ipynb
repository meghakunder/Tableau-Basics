**Q1. Explain what Tableau is and why it is preferred over traditional spreadsheet tools for data analysis.Also mention any two industries where Tableau is commonly used and explain one use case.**

**ANS:**

Tableau is a powerful data visualization and business intelligence software that helps users connect to data, analyze it, and create interactive dashboards and reports. It allows users to transform raw data into visual formats like charts, graphs, and maps, making insights easier to understand and act upon.



 **Why Tableau is preferred over traditional spreadsheet tools (like Excel)**

1. **Advanced Data Visualization**
   Tableau provides rich, interactive visuals (dashboards, heat maps, geographic maps) that are more dynamic and visually appealing than standard spreadsheet charts.

2. **Handles Large Data Efficiently**
   Unlike spreadsheets, which slow down with large datasets, Tableau can process millions of rows of data quickly.

3. **User-Friendly Interface**
   It uses drag-and-drop functionality, so even non-technical users can build complex visualizations without coding.

4. **Real-Time Data Connectivity**
   Tableau can connect to live data sources (databases, cloud services), enabling real-time analysis instead of static data.

5. **Interactive Dashboards**
   Users can filter, drill down, and explore data dynamically—something traditional spreadsheets offer only in limited ways.



**Industries where Tableau is commonly used**

1. **Healthcare**
2. **Finance (Banking & Financial Services)**


**Use Case Example (Healthcare)**

In the healthcare industry, Tableau is used to monitor patient data and hospital performance.

**Example use case:**
A hospital uses Tableau dashboards to track:

* Patient admissions and discharge rates
* Bed occupancy levels
* Disease trends over time

This helps hospital management:

* Allocate resources efficiently
* Predict patient inflow
* Improve quality of care

In summary, Tableau is preferred because it turns complex data into interactive, easy-to-understand insights, enabling faster and better decision-making compared to traditional spreadsheet tools.

**Q8. What is the difference between a Worksheet and a Dashboard in Tableau?**

**Why is a dashboard more useful for decision-makers?**

**ANS:**

**Worksheet**

A Worksheet is a single view or visualization

It shows one chart at a time (e.g., bar chart, line chart, map)

Used for individual analysis

 Example: Sales by Category chart

**Dashboard**

A Dashboard is a collection of multiple worksheets combined in one screen

It allows interaction between charts (filters, actions)

Used for overall analysis and storytelling

 Example: Sales dashboard with map + bar chart + trend line

 **Why Dashboard is more useful for decision-makers**

Provides complete picture in one place

Combines multiple insights (sales, profit, region, trends)

Enables quick comparison across different metrics

Interactive filters help users explore data easily

Saves time — no need to switch between sheets

**Q9.What is Tableau Prep, and why is it important in real-world data projects?**
**Then explain how you would:**

**Remove null values**

**Remove duplicate records**

**Change incorrect data types (Answer in steps, no tool access required.**

**ANS:**
Tableau Prep is a data preparation tool used to clean, transform, and combine data before analysis in Tableau.

**Why is it important in real-world projects?**

Real-world data is often:

Incomplete (null values)

Duplicated

Inconsistent formats

Tableau Prep helps to:

Improve data quality and accuracy

Save time during analysis

Ensure reliable insights and decisions

Without proper data preparation, analysis results can be misleading or incorrect.

1. Remove Null Values (Steps)

Open data in Tableau Prep

Go to a Clean Step

Select the column with null values

Click on “…” (More options)

Choose:

Filter → Exclude Null Values

 This removes all rows where the selected field is null

2. Remove Duplicate Records (Steps)

Add a Clean Step

Select all relevant columns

Click on “…” menu

Choose:

Remove Duplicates

Tableau Prep keeps only unique records

3. Change Incorrect Data Types (Steps)

Go to a Clean Step

Locate the column with incorrect data type

Click on the data type icon (e.g., ABC, #, calendar)

Select correct type:

String → Number

String → Date

etc.

 Ensures proper calculations and analysis
